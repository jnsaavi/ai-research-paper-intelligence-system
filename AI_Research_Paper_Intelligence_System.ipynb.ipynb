Custom Query Entity Classification
To categorize specific keywords in your query, we'll define a set of dictionaries mapping keywords to their respective types (e.g., 'framework', 'programming language', 'organization', 'model'). Then, we'll create a function to process your query and identify these custom entities.

```python
import re

frameworks = ['fast api', 'tensorflow', 'keras', 'scikit-learn', 'streamlit', 'django', 'flask', 'pytorch']
programming_languages = ['python', 'r', 'java', 'c++', 'javascript', 'go']
organizations = ['google', 'microsoft', 'ibm', 'facebook', 'amazon', 'apple']
models = ['bert', 'gpt', 'transformer', 'resnet', 'yolo', 'efficientnet', 'clip', 'dalle', 'llama', 'mistral', 'roberta']

def classify_query_entities(query):
    query_lower = query.lower()
    found_entities = {}

    # Check for frameworks
    for fw in frameworks:
        if re.search(r'\b' + re.escape(fw) + r'\b', query_lower):
            found_entities[fw] = 'Framework/Library'

    # Check for programming languages
    for lang in programming_languages:
        if re.search(r'\b' + re.escape(lang) + r'\b', query_lower):
            found_entities[lang] = 'Programming Language'

    # Check for organizations
    for org in organizations:
        if re.search(r'\b' + re.escape(org) + r'\b', query_lower):
            found_entities[org] = 'Organization'

    # Check for models
    for model_name in models:
        if re.search(r'\b' + re.escape(model_name) + r'\b', query_lower):
            found_entities[model_name] = 'Model'

    return found_entities
```

Let's test this custom entity classification function with your examples.

```python
print("Query: 'Python on medical dataset with fast api'")
entities1 = classify_query_entities("Python on medical dataset with fast api")
for entity, entity_type in entities1.items():
    print(f"- {entity}: {entity_type}")

print("\n" + "-" * 50 + "\n")

print("Query: 'research using pytorch and BERT from google'")
entities2 = classify_query_entities("research using pytorch and BERT from google")
for entity, entity_type in entities2.items():
    print(f"- {entity}: {entity_type}")

print("\n" + "-" * 50 + "\n")

print("Query: 'performance of GPT model in C++'")
entities3 = classify_query_entities("performance of GPT model in C++")
for entity, entity_type in entities3.items():
    print(f"- {entity}: {entity_type}")
```
